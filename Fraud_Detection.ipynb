```python
import pandas as pd  # We import the library that we are going to use
```

```python
df = pd.read_csv('creditcard.csv', index_col=False) #  We load the datasets
```

```python
df.head() # We visualize the first 5 rows
```

```python
df.Class.value_counts() #  We visualize the quantity of each class and we realize that the datasets is unbalanced
```

```python
df.hist(bins=30, figsize=(20,20)) # We see the distribution of the data of each column
```

```python
df.describe()
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
df['Amount'] = RobustScaler().fit_transform(df['Amount'].to_numpy().reshape(-1,1)) # We use RobustScaler to be able to work more easily with the Amount column
df['Amount'].hist()
```

```python
time = df['Time']
df['Time'] = (time - time.min()) / (time.max() - time.min()) # We scale the values in the Time column to be in a range from 0 to 1
df['Time'].hist()
```

```python
from sklearn.model_selection import train_test_split # We import train_test_split to separate training and test data
```

```python
X_train, X_test,y_train, y_test = train_test_split(df.iloc[:, :-1] ,df['Class'],random_state=104, test_size=0.25)
```

## We created the logistic regression model to classify fraud to no fraud


```python
from sklearn.linear_model import LogisticRegression 
logistic_model = LogisticRegression() 
logistic_model.fit(X_train, y_train)
logistic_model.score(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, logistic_model.predict(X_test), target_names=['Not Fraud', 'Fraud'])) # We evaluate the model
```

## We create the decision tree model to compare with the previous model

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth=2)
dtc.fit(X_train, y_train)
print(dtc.score(X_train, y_train))
print(classification_report(y_test, dtc.predict(X_test), target_names=['Not Fraud', 'Fraud']))
```

## We created RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=2,n_jobs=-1)
rf.fit(X_train, y_train)
print(classification_report(y_test, rf.predict(X_test), target_names=['Not Fraud', 'Fraud']))
```

## We created LinearSVC

```python
from sklearn.svm import LinearSVC
svc = LinearSVC(class_weight='balanced')
svc.fit(X_train, y_train)
print(classification_report(y_test, svc.predict(X_test), target_names=['Not Fraud', 'Fraud']))
```

## Finally we created a neural network to compare with the other models

```python

from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import InputLayer, Dense
from tensorflow.python.keras.callbacks import ModelCheckpoint
import tensorflow

modelo_nn = Sequential()
modelo_nn.add(InputLayer((X_train.shape[1],)))
modelo_nn.add(Dense(2, 'relu'))
modelo_nn.add(tensorflow.keras.layers.BatchNormalization())
modelo_nn.add(Dense(1, 'sigmoid'))

modelo_nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

modelo_nn.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=5)
```

```python
pred = (modelo_nn.predict(X_test).flatten() > 0.5).astype(int)
print(classification_report(y_test, pred, target_names=['Not Fraud', 'Fraud']))
```
